# Visualization of  the dataset

```python
data['is_canceled'].value_counts(normalize=True)
```

```python
plt.bar(["Not Cancelled","Cancelled"],data['is_canceled'].value_counts(),color='magenta',edgecolor='black')
plt.title("Is cancelled barplot")
plt.show()
```

```python
plt.figure(figsize=(8,4))
axis1 = sns.countplot(x='hotel', hue='is_canceled', data=data)
plt.title("Reservation status in different hotels")
plt.xlabel('hotels')
plt.ylabel("No. of reservation")
plt.legend("Not cancelled","Cancelled")
plt.show()
```

```python
data['hotel'].value_counts(normalize=True)
```

```python
City_hotel = data[data['hotel']=='City Hotel']
```

```python
Resort_hotel = data[data['hotel']=='Resort Hotel']
```

```python
City_hotel['is_canceled'].value_counts(normalize=True)
```

```python
Resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = Resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = City_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
resort_hotel
```

```python
plt.figure(figsize=(20,8))
plt.plot(resort_hotel.index,resort_hotel['adr'], label = "Resort rate")
plt.plot(city_hotel.index,city_hotel['adr'], label = "City rate")
plt.legend()
plt.title("Average daily rate fluctuation comparison")
plt.show()
```

```python
plt.figure(figsize=(16,8))
data['month'] = data['reservation_status_date'].dt.month
axis = sns.countplot(x='month', hue='is_canceled',data=data, palette ='bright')
legend_labels,_ = axis. get_legend_handles_labels()
axis.legend(bbox_to_anchor=(1,1))
plt.title("Cancelled amount in every month")
plt.xlabel('months')
plt.ylabel("No. of cancellation")
plt.legend(["Not cancelled","Cancelled"])
plt.show()
```
